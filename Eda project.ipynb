### Data Manipulation

#### Replacing blanks with 0 as tenure is 0 and no total charges are recorded.

```python
# Replace empty strings or strings with only spaces with 0
df["TotalCharges"] = df["TotalCharges"].replace(r'^\s*$', 0, regex=True)

# Convert to numeric
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
```

```python
print(df["TotalCharges"].head())
print(df["TotalCharges"].dtype)
```

```python
df.isnull().sum().sum()
```

### Percentage Of Missing Values

```python
round(df.isnull().sum()/len(df)*100)/2
```

```python
df['customerID'].duplicated().sum()
```

```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

#### converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

### Phase 2

### Data Visualizations

```python
ax=sns.countplot(x="Churn",data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(3,4))
ab=df.groupby("Churn").agg({'Churn':'count'})
plt.pie(ab['Churn'],labels= ab.index,autopct ="%1.2f%%")
plt.show()
ab
```

### 1. What percentage of customers have churned vs. stayed?


```python
plt.figure(figsize=(6,4))
sns.countplot(x="Churn", data=df)
plt.title("Churn Distribution")
plt.show()
```

### 2.Does churn vary by gender?


```python
plt.figure(figsize=(6,4))
sns.countplot(x="gender", hue="Churn", data=df)
plt.title("Churn by Gender")
plt.show()
```

### 3. Which contract type has the highest churn rate?


```python
plt.figure(figsize=(7,4))
sns.countplot(x="Contract", hue="Churn", data=df)
plt.title("Churn by Contract Type")
plt.xticks(rotation=30)
plt.show()
```

### 3. Do customers with higher monthly charges churn more?


```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges by Churn")
plt.show()
```

### 4.Does tenure affect churn probability?

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="tenure", hue="Churn", kde=True, bins=30)
plt.title("Tenure Distribution by Churn")
plt.show()
```

### 5.What’s the average total charges of churned vs. non-churned customers?

```python
plt.figure(figsize=(6,4))
df.groupby("Churn")["TotalCharges"].mean().plot(kind="bar", color=["steelblue","salmon"])
plt.title("Average Total Charges by Churn")
plt.ylabel("Average Total Charges")
plt.show()
```

### 6.Which numeric features are most correlated with churn?


```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

### 7.Is tenure strongly correlated with total charges?


```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="tenure", y="TotalCharges", hue="Churn", data=df)
plt.title("Tenure vs. Total Charges (colored by Churn)")
plt.show()
```

### 8.What’s the average total charges of churned vs. non-churned customers?


```python
plt.figure(figsize=(6,4))
df.groupby("Churn")["TotalCharges"].mean().plot(kind="bar", color=["steelblue","salmon"])
plt.title("Average Total Charges by Churn")
plt.ylabel("Average Total Charges")
plt.show()
```

### 9.Which numeric features are most correlated with churn?


```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

### 10.Are there extreme outliers in monthly charges?


```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["MonthlyCharges"])
plt.title("Boxplot of Monthly Charges")
plt.show()
```

### 11.Do outliers affect churn more (TotalCharges)?


```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Churn", y="TotalCharges", data=df)
plt.title("Total Charges by Churn (with outliers)")
plt.show()
```

### 11.Churn by Partner
