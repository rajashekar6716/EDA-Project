### Data Manipulation

#### Replacing blanks with 0 as tenure is 0 and no total charges are recorded.

```python
# Replace empty strings or strings with only spaces with 0
df["TotalCharges"] = df["TotalCharges"].replace(r'^\s*$', 0, regex=True)

# Convert to numeric
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
```

```python
print(df["TotalCharges"].head())
print(df["TotalCharges"].dtype)
```

```python
df.isnull().sum().sum()
```

### Percentage Of Missing Values

```python
round(df.isnull().sum()/len(df)*100)/2
```

```python
df['customerID'].duplicated().sum()
```

```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

#### converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

### Phase 2

### Data Visualizations

```python
# Churn distribution:
ax=sns.countplot(x="Churn",data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(3,4))
ab=df.groupby("Churn").agg({'Churn':'count'})
plt.pie(ab['Churn'],labels= ab.index,autopct ="%1.2f%%")
plt.show()
ab
```

### Univariate Analysis

```python
# Distribution of Monthly Charges
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
sns.histplot(df['MonthlyCharges'], kde=True, bins=30)
plt.title('Distribution of Monthly Charges')
```

### 📊 Insight:

-Most customers have monthly charges between 20–100 USD.


```python
# Count of Churn (Yes/No)
plt.subplot(1,2,2)
sns.countplot(x='Churn', data=df)
plt.title('Churn Count')
plt.show()
```


### 📊 Insight:

-About 26% of customers churn.



```python
# Distribution of Tenure
plt.figure(figsize=(10,5))
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title('Distribution of Customer Tenure')
plt.show()

```

### 📊 Insight:

-Most customers have tenure below 20 months → High churn risk in early stage.


### Bivariate Analysis

```python
# Tenure vs Monthly Charges by Churn
plt.figure(figsize=(14,6))
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=df, alpha=0.6)
plt.title('Tenure vs Monthly Charges (Churn highlighted)')
plt.show()
```

###  📊 Insight:

Churn is higher in month-to-month contracts than in 1-year or 2-year contracts.



```python
# Contract type vs Churn
plt.figure(figsize=(10,6))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Contract Type vs Churn')
plt.show()
```

### 📊 Insight:
Lower tenure customers are more likely to churn.



```python
# Boxplot: MonthlyCharges by Churn
plt.figure(figsize=(10,5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Monthly Charges Distribution by Churn Status')
plt.show()
```

### 📊 Insight:
Customers who pay higher monthly charges are more likely to churn.

```python
# Gender vs Churn
plt.figure(figsize=(8,5))
sns.countplot(x='gender', hue='Churn', data=df)
plt.title('Gender vs Churn')
plt.show()
```

### 📊 Insight:
No significant difference in churn between male and female customers.


### Multivariate Analysis

```python
# Pairplot
sns.pairplot(df[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']], hue='Churn', diag_kind='kde')
plt.suptitle('Pairplot of Numeric Features by Churn', y=1.02)
plt.show()
```

### 📊 Insight:
 Churned customers tend to have lower tenure and TotalCharges.


```python
# PaymentMethod vs Churn
plt.figure(figsize=(12,6))
sns.countplot(y='PaymentMethod', hue='Churn', data=df)
plt.title('Churn Distribution by Payment Method')
plt.show()
```

### 📊 Insight:
 Electronic Check customers are more likely to churn than those using other payment methods.


```python
# Contract and Internet Service vs Churn
sns.catplot(x='Contract', hue='Churn', col='InternetService', kind='count', data=df, height=5, aspect=1)
plt.suptitle('Impact of Contract and Internet Service on Churn', y=1.02)
plt.show()
```

### 📊 Insight:


Fiber optic users with month-to-month contracts are most likely to churn
